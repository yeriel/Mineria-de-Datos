```python
def plot_corr_matrix(data, features=None, annot=True, size=(16,10)):
    # MatPlotLib setup
    fig = plt.figure(figsize=size)
    ax = fig.add_subplot(1,1,1)

    # Select features to correlate
    if features is None:
        corr = data.corr()
    else:
        corr = data[features].corr()

    # Use Seaborn to plot 
    ax = sns.heatmap(corr, annot=annot,
                    xticklabels=corr.columns,
                    yticklabels=corr.columns,
                    cmap="seismic",
                    vmin=-1, vmax=1)
    plt.title("Correlation Matrix", fontsize = 15)
    plt.show()
```

### 2.- Regresión Lineal 1D

Con la característica seleccionada, realice una regresión lineal regularizada con modelo lineal. Incluya tanto regresión Ridge como Lasso. Grafique los datos y la línea de regresión.

**RECOMENDACIONES**:
* Estandarice sus datos
* Elimine las muestras con datos faltantes.
* Cree funciones para realizar las pruebas más facilmente.

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.kernel_ridge import KernelRidge
from sklearn.preprocessing import StandardScaler

def regresion(alpha):
    X_train = n_train.values
    std_scaler = StandardScaler()
    std_scaler.fit(X_train)
    std_train = std_scaler.transform(X_train)
    std_train_df = pd.DataFrame(std_train, columns=n_train.columns)
    std_train_df = std_train_df.dropna()

    X = std_train_df[['1stFlrSF']]
    xvals = np.linspace(X.min(), X.max(),X.size)

    # Regresion con Ridge
    ridge = Ridge(alpha)
    ridge.fit(X,std_train_df['SalePrice'])
    Ridge_Y_prediction = ridge.predict(xvals)

    # Regresion con Lasso
    lasso = Lasso(alpha)
    lasso.fit(X,std_train_df['SalePrice'])
    Lasso_Y_prediction = lasso.predict(xvals)

    # Plot Regresion
    fig, ax = plt.subplots()
    std_train_df.plot.scatter(x='TotalBsmtSF', y='SalePrice', color='yellow', label='Data', alpha=0.5, ax=ax, title='alpha '+str(alpha))
    ax.plot(xvals, Ridge_Y_prediction, color='red', linewidth=2.0, label='Ridge Regression Line')
    ax.plot(xvals, Lasso_Y_prediction, color='blue', linewidth=2.0, label='Lasso Regression Line')
    plt.legend()
    plt.figure()
```

```python
Range = np.linspace(0.1,1,10)
for i in Range:
    regresion(i)
```

**PREGUNTA**: ¿Qué valores de `alpha` debemos utilizar para ver cambios en el comportamiento de la regresión tanto en Ridge como en Lasso? ¿Por qué cree que son tan distintos?

**RESPUESTA**: En la regresión lineal Lasso el parámetro de ajusta alpha controla la penalización de la norma L1. Entonces cuando alpha toma un valor sero es igual la uan regresión lineal tradicional y a medida que este aumenta más coeficientes de la regresión se ponderan a cero, es decir a medida que aumenta alpha aumenta el sesgo y a medida que disminuye  la varianza va aumentando. Entonce un buen rango para el parámetro alpha en esta regresión va entre cero y 1.

En Cuanto a la regresión Ridge entre más grande sea el valor alpha mayor será el tamaño de los coeficientes penalizado. La diferencia entre las dos regresiones se debe principalmente a el tipo de regulación que utiliza cada uno para Lasso la norma L1 y Ridge la norma L2
 

### 3.- Regresión en ND

Vuelva a utilizar todo el dataset, pero separe ahora el conjunto de entrenamiento en un conjunto de entrenamiento que contenga el 80% del original, y un conjunto de validación con el 20% del original (hint: `sklearn.model_selection`). Entrene ambos regresores regularizados para distintos valores de `alpha` (100 valores distintos), y calcule el error cuadrático medio de entrenamiento y el error del conjunto de validación. Con estos datos, gráfique la evolución del error de entrenamiento y validación dependiendo del peso de regularización. (hint: no calcule el error usted, ocupe `sklearn.metrics`!). Haga lo mismo para el regresor `KernelRidge` con parámetro `kernel='rbf'`.

**RECOMENDACIONES**:
* Estandarice el set de entramiento y validación. Recuerde ocupar la misma tranformación para el testing.
* Elimine las muestras con datos faltantes y verifique que los conjuntos de entrenamiento y testing sean congruentes.
* Cree funciones para realizar las pruebas más facilmente.

```python
def std_df(df):
    n_train = df.select_dtypes(include=np.number)
    X_train = n_train.values
    std_scaler = StandardScaler()
    std_scaler.fit(X_train)
    std_train = std_scaler.transform(X_train)
    std_train_df = pd.DataFrame(std_train, columns=n_train.columns)
    std_train_df = std_train_df.dropna()
    return std_train_df
```

```python
train_df = pd.read_csv('train.csv')

#std df train 
std_train_df = std_df(train_df)
```

```python
def RIDGE(X_train,Y_train,alpha):
    # Regresion con Ridge
    ridge_regressor = Ridge(alpha)
    ridge_regressor.fit(X_train, Y_train)
    Ridge_Y_prediction = ridge_regressor.predict(X_train)
    return  Ridge_Y_prediction
```

```python
def LASSO(X_train,Y_train,alpha):
    lasso = Lasso(alpha)
    lasso.fit(X_train,Y_train)
    Lasso_Y_prediction = lasso.predict(X_train)
    return Lasso_Y_prediction
```

```python
def KERENLRIDGE(X_train,Y_train,alpha):
    # Regresion con KernelRidge
    K_ridge_regressor = KernelRidge(kernel='rbf',alpha=alpha)
    K_ridge_regressor.fit(X_train, Y_train)
    KRidge_Y_prediction = K_ridge_regressor.predict(X_train)
    return KRidge_Y_prediction
```

```python
def plotTrainValid(alpha, yTrain, yValid, title):
    fig, ax = plt.subplots()
    ax.plot(alpha, yTrain, color='red', linewidth=2.0, label=title+" Training MSE")
    ax.plot(alpha, yValid, color='blue', linewidth=2.0, label=title+" Validation MSE")

    plt.ylabel(title)
    plt.xlabel("alpha")
    plt.title(title + " MSE vs alpha")
    plt.legend()
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

#df train 0.8 & validation0.2
std_train, std_val, std_y_train, std_y_val = train_test_split(std_train_df, std_train_df['SalePrice'], train_size=0.8)
print('shape std_train ' + str(std_train.shape))
print('shape std_val ' + str(std_val.shape))
```

```python
#error para Ridge
msrRidgeTrain = []
msrRidgeVal = []
Range = np.linspace(0,10**5,100)

for i in Range:
    #Conjunto train
    predict_ridge_train = RIDGE(std_train,std_y_train,i)
    msrT = mean_squared_error(std_y_train,predict_ridge_train)
    msrRidgeTrain.append(msrT)
    
    #Conjunto Val
    predict_ridge_val = RIDGE(std_val,std_y_val,i)
    msrV = mean_squared_error(std_y_val,predict_ridge_val)
    msrRidgeVal.append(msrV)

plotTrainValid(Range,msrRidgeTrain,msrRidgeVal,'Ridge')
```

```python
#error para Lasso
msrLassoTrain = []
msrLassoVal = []
Range = np.linspace(10**-6,2,100)

for i in Range:
    #Conjunto train
    predict_lasso_train = LASSO(std_train,std_y_train,i)
    msrT = mean_squared_error(std_y_train,predict_lasso_train)
    msrLassoTrain.append(msrT)
    
    #Conjunto Val
    predict_lasso_val = LASSO(std_val,std_y_val,i)
    msrV = mean_squared_error(std_y_val,predict_lasso_val)
    msrLassoVal.append(msrV)

plotTrainValid(Range,msrLassoTrain,msrLassoVal,'Lasso')
```

```python
#error para KernelRigde
msrKrTrain = []
msrKrVal = []
Range = np.linspace(0.1,1,100)

for i in Range:
    #Conjunto train
    predict_kr_train = KERENLRIDGE(std_train,std_y_train,i)
    msrT = mean_squared_error(std_y_train,predict_kr_train)
    msrKrTrain.append(msrT)
    
    #Conjunto Val
    predict_kr_val = KERENLRIDGE(std_val,std_y_val,i)
    msrV = mean_squared_error(std_y_val,predict_kr_val)
    msrKrVal.append(msrV)

plotTrainValid(Range,msrKrTrain,msrKrVal,'KernelRidge ')
```

**PREGUNTA**: ¿Qué comportamiento observa en la evolución del error? Si no observa ningún comportamiento, cambie los valores de `alpha`.

**RESPUESTA**: Se puede apreciar que el error va en aumento tanto el de train como el de validación hasta un punto en el que el error de validación se hace mayor que el de entrenamiento. En cuanto al caso de KernelRidge el error de validación no sobrepasa el el error de entrenamiento para el rango de valores de alpha que se definió. El mejor regresos seria KernelRidge

### 4 - Predicción
Estime el precio de las casas en el conjunto de testing utilizando el mejor de los regresores con su mejor `alpha`. Grafíque la dispersión de cada característica de la sección 1 versus el precio de venta, tanto para los datos de training y testing en distintos colores.

```python
def KERENLRIDGE(X_train,Y_train,X_test,alpha):
    # Regresion con KernelRidge
    K_ridge_regressor = KernelRidge(alpha=alpha)
    K_ridge_regressor.fit(X_train, Y_train)
    KRidge_Y_prediction = K_ridge_regressor.predict(X_test)
    return KRidge_Y_prediction
```

```python
test_df = pd.read_csv('test.csv')
train_df = pd.read_csv('train.csv')

n_test_df = test_df.select_dtypes(include=np.number).dropna()
n_train_df = train_df.select_dtypes(include=np.number).dropna()

Y = n_train_df['SalePrice']
X = n_train_df.drop(columns='SalePrice')
alphaIdeal = 0.15

predictedPrices = KERENLRIDGE(X,Y,n_test_df,alphaIdeal)
predicted_df = n_test_df.copy()
predicted_df['PredictedPrices'] = predictedPrices
```

```python
for i in range(0,len(feat)-1):
    fig=plt.figure(figsize=(8,5))
    ax=fig.add_subplot(1,1,1)
    ax.scatter(predicted_df[feat[i]], predicted_df['PredictedPrices'], color='y', label='predicted')
    ax.scatter(n_train_df[feat[i]], n_train_df['SalePrice'], color='b', label='test', alpha=0.1)
    ax.set_xlabel(feat[i])
    ax.set_ylabel('Prices')
    ax.legend()
    plt.show()
```
