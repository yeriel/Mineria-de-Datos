## 1.- Carga y Correlación de Datos

En el aula encontrará el archivo ``fetal_health.csv`` que contiene 2126 registros de **cardiotocografías**, las cuales fueron clasificadas por un panel de expertos (obstetras) en tres clases:

* Normal = 1
* Sospechoso = 2

Las cardiotocografías son una forma simple y económica de monitorear la salud fetal, permitiendo a los profesionales de la salud tomar acciones para prevenir muertes tanto los bebés como las madres. El instrumento como tal funciona enviando pulsos ultrasonoros y midiendo su respuesta, dando nos información del ritmo cardiaco fetal(FHR), movimiento fetal, contracciones uterinas, entre otros.

Cargue el dataset indicado en un DataFrame y muestre gráficamente su matriz de correlación. Luego copie la columna `fetal health` en una nueva variable, y elimínela de su DataFrame (o cree uno nuevo sin esa columna). 

```python
import seaborn as sns
import matplotlib.pyplot as plt
def plot_corr_matrix(data,features=None,annot=True,s=(12,8)):
    fig= plt.figure(figsize=s)
    ax= fig.add_subplot(111)
    if features is None:
        corr = data.corr()
    else:
        corr = data[features].corr()
    ax= sns.heatmap(corr,annot=annot,
        xticklabels=corr.columns,
    yticklabels=corr.columns, cmap="seismic",vmin=-1,vmax=1)
    plt.title("Correlation Matrix", fontsize = 15)
    plt.show()
```

```python
# Escriba su código aquí
import pandas as pd
table_fh = pd.read_csv("fetal_health.csv")
plot_corr_matrix(table_fh)
```

```python
fetal_health = table_fh['fetal_health']
del table_fh['fetal_health']
table_fh
```

## 2.- Visualización de Datos
Primero estandarice sus datos (StandardScaler) y trabaje desde ahora con esos datos para las siguientes preguntas.
Para visualizar los datos, utilice la técnica T-SNE en dos dimensiones. Coloree cada examen utilizando la columna ``fetal_health``que separó.

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

num_feat = ['baseline value', 'accelerations', 'fetal_movement',
       'uterine_contractions', 'light_decelerations', 'severe_decelerations',
       'prolongued_decelerations', 'abnormal_short_term_variability',
       'mean_value_of_short_term_variability',
       'percentage_of_time_with_abnormal_long_term_variability',
       'mean_value_of_long_term_variability', 'histogram_width',
       'histogram_min', 'histogram_max', 'histogram_number_of_peaks',
       'histogram_number_of_zeroes', 'histogram_mode', 'histogram_mean',
       'histogram_median', 'histogram_variance', 'histogram_tendency']

scaler = StandardScaler().fit(table_fh[num_feat])
```

```python
from sklearn.manifold import TSNE
def plot_tsne(data, feats, label, perplxty=20):
    tsne_model = TSNE(n_components = 2, init = 'pca', perplexity=perplxty)
    tracks_tsne = tsne_model.fit_transform(data[feats])

    fig, ax = plt.subplots(figsize = (10, 10), sharex = True)
    for playlist in np.unique(label):
        mask = (label == playlist)
        ax.scatter(tracks_tsne[mask][:, 0], tracks_tsne[mask][:, 1], label = playlist, alpha = 0.5)
    ax.set_xlabel("t-SNE 1")
    ax.set_ylabel("t-SNE 2")
    ax.legend()

plot_tsne(table_fh, num_feat, fetal_health)
```

```python
plot_tsne(table_fh, num_feat, fetal_health,50)
```

```python
plot_tsne(table_fh, num_feat, fetal_health, 10)
```

## 3.- Reducción de dimensionalidad
Realice un PCA lineal sobre el DataFrame estandarizado (**sin la columna** ``fetal_health``), utilizando la misma cantidad de componentes que características existentes en la tabla. Grafique la varianza explicada por las componentes del PCA, y seleccione las primeras $n$ características que expliquen aproximadamente el 90% de la varianza del dataset. Por último, grafique la matriz de correlaciones de los $n$ componentes seleccionados y realice un T-SNE con los mismos parámetros de la sección anterior sobre estos $n$ componentes (elija la perplejidad que más le acomode).


```python
# Escriba su código aquí
scaler = StandardScaler().fit(table_fh[num_feat])
stdtab = pd.DataFrame(scaler.transform(table_fh[num_feat]))
n = 11
sklearn_pca = PCA(n_components=n,random_state=1)
xpca = sklearn_pca.fit_transform(stdtab)
varx=sklearn_pca.explained_variance_ratio_
#Grafica la varianza acumulada
plt.plot(np.arange(1,n+1),varx.cumsum())
plt.ylabel("% of variance")
plt.xlabel("components")
```

```python
varx.sum()*100
```

```python
n = 11
ind = []
for i in range(n):
    ind.append("PC"+str(i+1))
table_pca = pd.DataFrame(xpca, columns=ind)
plot_corr_matrix(table_pca)
```

```python
plot_tsne(table_pca, ind, fetal_health,10)
```

## 4.- Clustering 
Realice un clustering utilizando KMeans con 3 clases (ya que hay 3 clases de ``fetal_health``) **sobre los $n$ componentes del PCA**. Visualice los resultados de este clustering aplicando la técnica T-SNE al igual que la pregunta anterior. Repita el mismo procedimiento de clustering y visualización utilizando una **GaussianMixture** (mezcla de Gaussianas) nuevamente con solo 3 mezclas.

```python
from sklearn.cluster import KMeans
n = 11
ind = []
for i in range(n):
    ind.append("PC"+str(i+1))
table_pca = pd.DataFrame(xpca, columns=ind)

km = KMeans(3).fit(table_pca[ind]) 
table_pca['cluster']=km.labels_
plot_tsne(table_pca, ind, table_pca['cluster'])
```

```python
from sklearn.mixture import GaussianMixture

ind = []
for i in range(n):
    ind.append("PC"+str(i+1))
table_pca = pd.DataFrame(xpca, columns=ind)

gmm = GaussianMixture(n_components=3).fit(table_pca)
table_pca['cluster']=gmm.predict(table_pca)
plot_tsne(table_pca, ind, table_pca['cluster'])
```
